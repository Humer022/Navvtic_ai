```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression



np.random.seed(0)

x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x+np.random.rand(100, 1)

# 2. Train a Linear Regression model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

# 3. Plot the scatter plot for original data
plt.scatter(x,y,color='blue',label='Actual data' )

# 4. Plot the predicted points (red circles)
plt.scatter(y,y_pred)
# , facecolors='none', label='Predicted Data')
# plt.scatter(x, y_pred,facecolors='none', edgecolors='red', label='Predicted Data')
plt.title('Regression: Actual vs Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#questin no 2
import seaborn as sns
heights = np.random.normal(160, 10, 100)  

# Univariate Plot: Histogram
plt.figure(figsize=(8,5))
sns.histplot(heights, kde=True, color='skyblue')
plt.title('Univariate Plot: Heights Histogram')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python

weights = 0.5 * heights + np.random.normal(30, 5, 100)  

# Bivariate Plot: Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(heights, weights, color='green')
plt.title('Bivariate Plot: Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
